```python
### segement types
type_of_segment= df["Segment"].unique()
print(type_of_segment)
```

```python
## no of customers in each segment
no_of_customers= df["Segment"].value_counts().reset_index()
no_of_customers= no_of_customers.rename(columns={'Segment':'customers_type', 'count':'total_customers'})

print(no_of_customers)
```

```python
## Ploting pie chart
plt.pie(no_of_customers['total_customers'], labels=no_of_customers['customers_type'], autopct="%1.1f%%")
plt.title("Distribution of customers")
plt.show()
```

```python
##  Customers and Salessale
sales_per_category = df.groupby('Segment')['Sales'].sum().reset_index()

sales_per_category= sales_per_category.rename(columns={'Segment':'customers_type', 'count':'Sales'})
print(sales_per_category)
```

```python
## Ploting pie chart
plt.pie(sales_per_category['Sales'], labels=sales_per_category['customers_type'], autopct="%1.1f%%")
plt.title("Sales per customer category")
plt.show()
```

```python
# Plotting by bar graph
plt.bar(sales_per_category["customers_type"],sales_per_category['Sales'])
plt.title("Sales per customer category")
plt.xlabel("Customer Type")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Grouping data according to 'customerid, name, segment and calculate the freq of their orders
customer_order_freq=df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Order ID'].count().reset_index()
# Rename the colunm
customer_order_freq.rename(columns={'Order ID':'Total_orders'}, inplace=True)
# Identify the repeat customers

repeat_customers = customer_order_freq[customer_order_freq['Total_orders']>=1]

# sort the repaet customers in descending order
sort_repeat_customers= repeat_customers.sort_values(by='Total_orders', ascending=False)

print(sort_repeat_customers.head(10).reset_index(drop=True))
```

```python
# Grouping data according to 'customerid, name, sales
customer_sales=df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Sales'].sum().reset_index()
# sort in desc order
top_spenders= customer_sales.sort_values(by='Sales', ascending=False)
print(top_spenders.head(10).reset_index(drop=True))
```

## Sales Trend Analysis

```python
# convert order date in to datetime format
df['Order Date']= pd.to_datetime(df['Order Date'], dayfirst= True)

#Grouping by year and summing the sales per year
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()

# Setting new index and remaing columns
yearly_sales =yearly_sales.reset_index()
yearly_sales= yearly_sales.rename(columns= {'Order Date': 'Year',
    'Sales': 'Total Sales'})

print(yearly_sales)
```

```python
# Plooting bar graph on this subcategory vs sales
plt.bar(yearly_sales['Year'], yearly_sales['Total Sales'])

# label
plt.title('Yearly sales')
plt.xlabel('Years')
plt.ylabel('Total Sales')
plt.xticks(rotation=70)
plt.show()
```

```python
# Plooting line graph on this subcategory vs sales
plt.plot(yearly_sales['Year'], yearly_sales['Total Sales'], marker='o', linestyle='--')

# label
plt.title('Yearly sales')
plt.xlabel('Years')
plt.ylabel('Total Sales')
plt.xticks(rotation=70)
plt.show()
```

# Quarterly Sales


```python
# convert order date in to datetime format
df['Order Date']= pd.to_datetime(df['Order Date'], dayfirst= True)

#Grouping by year and summing the sales per quarter
year_sales = df[df['Order Date'].dt.year == 2018]

quarterly_sales =year_sales.resample('QE',on='Order Date') ['Sales'].sum()

# Setting new index and remaing columns
quarterly_sales =quarterly_sales.reset_index()
quarterly_sales= quarterly_sales.rename(columns= {'Order Date': 'Quarter',
    'Sales': 'Total Sales'})

print(quarterly_sales)
```

```python
# Plooting bar graph on this subcategory vs sales
plt.bar(quarterly_sales['Quarter'], quarterly_sales['Total Sales'])

# label
plt.title('Quarterly sales')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(rotation=70)
plt.show()
```

```python
# Plooting line graph on this subcategory vs sales
plt.plot(quarterly_sales['Quarter'], quarterly_sales['Total Sales'], marker='*', linestyle='--')

# label
plt.title('Quarterly sales')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(rotation=70)
plt.show()
```

## Monthly sales

```python
# convert order date in to datetime format
df['Order Date']= pd.to_datetime(df['Order Date'], dayfirst= True)

#Grouping by year and summing the sales per quarter
year_sales = df[df['Order Date'].dt.year == 2018]

monthly_sales =year_sales.resample('ME',on='Order Date') ['Sales'].sum()

# Setting new index and remaing columns
monthly_sales =monthly_sales.reset_index()
monthly_sales= monthly_sales.rename(columns= {'Order Date': 'Monthly',
    'Sales': 'Total Sales'})

print(monthly_sales)
```

```python
# Plooting line graph on this subcategory vs sales
plt.plot(monthly_sales['Monthly'], monthly_sales['Total Sales'], marker='*', linestyle='--')

# label
plt.title('Monthly sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=70)
plt.show()
```
